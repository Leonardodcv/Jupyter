```python
#parallel sirve para poder paralelizar el proceso
#realizado entre varios nucleos del procesador
import pandas as pd
import numpy as np
from joblib import Parallel, delayed
df = pd.read_csv("D:/Python/achieves/2008.csv", nrows=100000)
```

```python
def_sub = df[["CarrierDelay","WeatherDelay","NASDelay","SecurityDelay","LateAircraftDelay"]]
def_sub.head(10)
```

```python
def retraso_maximo(fila):
    if not np.isnan(fila).any():
        names=["CarrierDelay","WeatherDelay","NASDelay","SecurityDelay","LateAircraftDelay"]
        return names[fila.index(max(fila))]
    else:
        return "None"
    
        
```

```python
results=[]
for fila in def_sub.values.tolist():
    results.append(retraso_maximo(fila))
```

```python
results
```

```python
result=Parallel(n_jobs=2, backend = "multiprocessing")(
    map(delayed(retraso_maximo),)
)
```
