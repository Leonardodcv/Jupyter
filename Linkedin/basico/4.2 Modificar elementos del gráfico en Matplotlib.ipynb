```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(0)
df = pd.read_csv("D:/Python/achieves/2008.csv", nrows=100000)
df = df.sample(frac=1).head(100)
```

```python
plt.scatter(x = df.DayofMonth, y = df.ArrDelay, s = df.Distance)
```

```python
plt.scatter(x = df.DayofMonth, y = df.ArrDelay, s = df.Distance, alpha =0.3)
```

```python
plt.scatter(x = df.DayofMonth, y = df.ArrDelay, s = df.Distance, alpha =0.3, c = df.DayOfWeek.isin([6,7]))
```

```python
plt.scatter(x = df.DayofMonth, y = df.ArrDelay, s = df.Distance, alpha =0.3, c = df.DayOfWeek.isin([6,7]))
plt.title("Retrasos en vuelos")
plt.xlabel("Dia del Mes")
plt.ylabel("Retraso en la llegada")
plt.ylim([0,150])
plt.xticks([0,15,30])
plt.text(x=25,y=120,s="Mi Vuelo")
plt.show()
```
