```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("D:/Python/achieves/2008.csv", nrows=100000)
#arreglo con un conteo las veces que aparece un dia a la semana
data = np.unique(df.DayOfWeek,return_counts=True)
labels=["Lun","Mar", "Mie", "Jue","Vie","Sab","Dom"]
data
```

```python
plt.pie(x = data[1],
       labels=labels,
       radius = 1.5,
       colors = sns.color_palette("hls", 7),
        explode = (0,0,0,0.2,0,0,0.1),
       startangle = 90,
       autopct = "%1.1f%%",
       labeldistance= 1.1)
plt.legend(loc="upper left", labels =labels)
plt.show()
```

```python
plt = sns.barplot(x=labels, y = data[1])
plt.set(xlabel = "Dia de la semana", ylabel="Numero de vuelos")
```
