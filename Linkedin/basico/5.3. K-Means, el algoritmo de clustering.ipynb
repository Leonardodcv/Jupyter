```python
#la idea del algoritmo k-meanssepara los grupos en 
#en un numero predefinido con una varianza parecida
#otra cosa que hace el algoritmo es generar grupos
#disjuntos lo cual se consigue asignando cada uno 
#de los puntos a los centroides mas cercanos.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv("D:/Python/achieves/2008.csv", nrows = 1e5)
#elimina los casos faltantes de las dos colemnas entre []
newdf = df[["AirTime","DepDelay"]].dropna()
```

```python
newdf = df[["AirTime","Distance","TaxiOut","ArrDelay","DepDelay"]].dropna()
```

```python
#n_clusters es el parametro principal y es el numero
#de centros que se requieren
#ramdon_state servira para replicar los resultados
#n_jobs le dice a la computadora que use tos los 
#nucleos del procesador.
#fit(newdf) servira para ajustar los datos en base a
#las columnas que hay en newdf
kmeans = KMeans(n_clusters = 4, random_state=0).fit(newdf)
kmeans.labels_
```

```python
np.unique(kmeans.labels_,return_counts=True)
```

```python
import matplotlib.pyplot as plt
plt.scatter(newdf["AirTime"], newdf["DepDelay"], c = kmeans.labels_)
plt.show()
```

```python
kmeans.cluster_centers_
```
